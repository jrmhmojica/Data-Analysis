# Stack Overflow 2022 Survey Analysis
This is an analysis of the 2022 Stack Overflow Survey. The survey has more than 70,000 respondents but I only chose to analyze the submissions of respondents from the Philippines.

```python
import pandas as pd
dataset = pd.read_csv("data/survey_results_public.csv")
dataset
```

```python
schema = pd.read_csv("data/survey_results_schema.csv")
schema
```

```python
list(dataset.columns)
```

```python
survey = dataset.copy()
survey["Country"].value_counts()
```

## Filtering the dataset
Getting only the submissions of respondents from the Philippines

```python
survey_ph = survey[survey["Country"] == "Philippines"]
pd.set_option('display.max_columns', None)
survey_ph
```

There are 366 respondents from the Philippines in the Stack Overflow 2022 Survey

## Basic Information


```python
survey_ph["MainBranch"].value_counts()
```

```python
survey_ph["MainBranch"].value_counts(normalize=True)
```

222 people or ~60% of respondents from the Philippines are developers by profession.

```python
survey_ph["Age"].value_counts()
```

```python
survey_ph["Age"].value_counts(normalize=True)
```

People aged 25 - 34 years old represents almost 46% of the respondents, and people aged 18 - 24 years old represents 32%. 10% of the respondents are under 18 years of age while the remaining are aged 35 and above.

```python
survey_ph["Gender"].value_counts()
```

```python
survey_ph["Gender"].value_counts(normalize=True)
```

Majority of the respondents from the Philippines are men. 88% are men, 7% are women, and the remaining are part of the LGBT community.

```python
sph_employment = survey_ph["Employment"].str.split(";", expand=True)
sph_employment.stack().value_counts()
```
